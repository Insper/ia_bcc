```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
```

```python
df = pd.read_csv('./data/dados_treinamento.csv')
df['Rewards_Avg'] = df['Rewards'].rolling(window=100).mean()
df['Actions_Avg'] = df['Actions'].rolling(window=100).mean()
df.head()
```

```python
sns.lineplot(x='Episode', y='Rewards_Avg', data=df)
plt.title('Rewards per Episode')
plt.show()
```

```python
sns.lineplot(x='Episode', y='Epsilon', data=df)
plt.title('Epsilon per Episode')
plt.show()
```

```python
sns.lineplot(x='Episode', y='Actions_Avg', data=df)
plt.title('Actions per Episode')
plt.show()
```
