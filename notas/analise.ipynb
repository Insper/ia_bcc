```python
import pandas as pd 
import seaborn as sns
file_path = "notas_black.csv"  
data = pd.read_csv(file_path)
data.head()
```

```python
data.iloc[:, 4:] = data.iloc[:, 4:].replace(',', '.', regex=True)
for col in data.columns[4:]:
    data[col] = pd.to_numeric(data[col])
```
