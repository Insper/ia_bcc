```python
import pandas as pd 
import seaborn as sns
#file_path = "notas_black.csv" # 2025-1
file_path = "2025_2_notas.csv" # 2025-2  
data = pd.read_csv(file_path)
data.head()
```

```python
data.iloc[:, 6:] = data.iloc[:, 6:].replace(',', '.', regex=True)
for col in data.columns[6:]:
    data[col] = pd.to_numeric(data[col])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seleciona as colunas de interesse
columns = ['AI [Total de pontos: 10 Pontuação] |428124',
           'AF [Total de pontos: 10 Pontuação] |428128']

# Estilo do Seaborn
sns.set_theme(style="whitegrid", context="notebook")

# Converte para formato longo para o Seaborn
long_df = data[columns].melt(var_name='Avaliação', value_name='Pontuação')

plt.figure(figsize=(10, 6))
ax = sns.boxplot(data=long_df, x='Avaliação', y='Pontuação', palette='Set2')
#sns.stripplot(data=long_df, x='Avaliação', y='Pontuação', color='black', alpha=0.5, size=3, jitter=0.2)

ax.set_title('Boxplots entre AI e AF', fontsize=14)
ax.set_xlabel('Avaliação')
ax.set_ylabel('Pontuação')

plt.tight_layout()
plt.show()
```

```python
import re
import seaborn as sns
import matplotlib.pyplot as plt

# Detecta a coluna de nota geral; fallback: soma de AI + AF
candidate_cols = [
    c for c in data.columns[6:]
    if re.search(r'\b(geral|final|total)\b', c, flags=re.I)
    and not c.strip().startswith(('AI ', 'AF '))
]
if candidate_cols:
    nota_col = candidate_cols[0]
else:
    ai_cols = [c for c in data.columns if c.strip().startswith('AI ')]
    af_cols = [c for c in data.columns if c.strip().startswith('AF ')]
    data['Nota Geral (AI+AF)'] = data[ai_cols + af_cols].sum(axis=1)
    nota_col = 'Nota Geral (AI+AF)'

# Quantidade de notas abaixo de 5
vals = data[nota_col].dropna()
qtd_abaixo_5 = int((vals < 5).sum())

# Plot
sns.set_theme(style="whitegrid", context="notebook")
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(data=data, x=nota_col, bins=20, color='#4C72B0', edgecolor='white', ax=ax)

ax.axvline(5, color='red', linestyle='--', linewidth=1, label='Limite 5')
ax.set_title('Histograma da Nota Geral', fontsize=14)
ax.set_xlabel('Nota Geral')
ax.set_ylabel('Frequência')
ax.legend()

caption = f'Notas abaixo de 5: {qtd_abaixo_5} (N={len(vals)})'
fig.text(0.5, -0.05, caption, ha='center', va='top', fontsize=10)

plt.tight_layout()
plt.show()
```
